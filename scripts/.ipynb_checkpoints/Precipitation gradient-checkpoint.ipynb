```python
%config Completer.use_jedi = False
import numpy as np
from osgeo import gdal
from osgeo import gdal_array
from osgeo import osr
import matplotlib.pylab as plt
```

# Load precipitation data

```python
img_ds=gdal.Open('D:/Proyectos_GitHub/Bartlet-Lewis_Regionalization/output/CV_maps/average_rainfall_deph_storm.tif')
average_deph_pp=img_ds.GetRasterBand(1).ReadAsArray()
```

```python
average_deph_pp[average_deph_pp<0]=0
```

```python
plt.figure(figsize=(15,10))
plt.imshow(average_deph_pp)
plt.show()
```

```python
nx,ny=average_deph_pp.shape
xnder=np.zeros((nx,ny))  # numerical derivative
ynder=np.zeros((nx,ny))
dx=11100 # km
dy=dx
```

```python
for j in range(0,ny):
    for i in range(1, nx - 1):
        xnder[i,j]=(average_deph_pp[i+1,j]-average_deph_pp[i-1,j])/(2*dx)
        
        
for i in range(0,nx):
    for j in range(1, ny - 1):
        ynder[i,j]=(average_deph_pp[i,j+1]-average_deph_pp[i,j-1])/(2*dy)
```

```python
plt.figure(figsize=(15,10))
plt.imshow(ynder)
plt.show()
```

```python
x = np.arange(0,nx,1)
y = np.arange(0,ny,1)
X, Y = np.meshgrid(x, y)
```

```python
fig, ax = plt.subplots(figsize=(7,7))
ax.quiver(X,Y,xnder,ynder)
ax.xaxis.set_ticks([])
ax.yaxis.set_ticks([])
ax.set_aspect('equal')

plt.show()
```
